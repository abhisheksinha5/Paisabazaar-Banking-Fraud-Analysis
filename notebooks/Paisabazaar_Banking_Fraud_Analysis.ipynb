## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
# Making a opy of the cleaned dataset
df_wrangled = df.copy()
```

```python
# Checking for missing values in the dataset
print("Missing Values Count:")
df_wrangled.isnull().sum()
```

After checking for **NaN values** and **null values** across all columns, it was observed that the dataset contains **no missing values**.

This confirms that the data is **complete**, **consistent**, and does not require any **imputation**, **data replacement**, or **row/column removal** before proceeding with further **exploratory data analysis (EDA)** and **modeling**.

```python
# 1. Age Group
df_wrangled['Age_Group'] = pd.cut(
    df_wrangled['Age'],
    bins=[0, 25, 35, 45, 55, 100],
    labels=['18-25', '26-35', '36-45', '46-55', '55+']
)

print("Age Group Distribution:")
display(df_wrangled['Age_Group'].value_counts())
```

```python
# 2. Income Category
df_wrangled['Income_Category'] = pd.qcut(
    df_wrangled['Annual_Income'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Very High']
)

print("\nIncome Category Distribution:")
display(df_wrangled['Income_Category'].value_counts())
```

```python
# 3. Credit Utilization Level
df_wrangled['Credit_Utilization_Level'] = pd.cut(
    df_wrangled['Credit_Utilization_Ratio'],
    bins=[0, 30, 60, 100],
    labels=['Low', 'Medium', 'High']
)

print("\nCredit Utilization Level Distribution:")
display(df_wrangled['Credit_Utilization_Level'].value_counts())
```

```python
# 4. Payment Discipline Flag
df_wrangled['Poor_Payment_Discipline'] = np.where(
    df_wrangled['Num_of_Delayed_Payment'] > 0, 'Yes', 'No'
)

print("\nPayment Discipline Distribution:")
display(df_wrangled['Poor_Payment_Discipline'].value_counts())
```

```python
# 5. Debt Burden Ratio
df_wrangled['Debt_Burden_Ratio'] = (
    df_wrangled['Total_EMI_per_month'] / df_wrangled['Monthly_Inhand_Salary']
)

print("\nDebt Burden Ratio Summary:")
display(df_wrangled['Debt_Burden_Ratio'].describe())
```

#### Chart - 12 SC-Params Risk Decision Matrix: Credit Score vs Delayed Payments

```python
# Chart - 12 visualization code

# Creating delayed payment bins
df['Delayed_Payment_Band'] = pd.cut(
    df['Num_of_Delayed_Payment'],
    bins=[-1, 5, 10, df['Num_of_Delayed_Payment'].max()],
    labels=['Low Delay', 'Medium Delay', 'High Delay']
)

# Creating pivot table for risk distribution
risk_matrix = pd.crosstab(
    df['Delayed_Payment_Band'],
    df['Credit_Score']
)

print("Risk Decision Matrix (Count):")
print(risk_matrix)
print(" ")

# Visualization: Risk Decision Matrix Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(
    risk_matrix,
    annot=True,
    fmt='d',
    cmap='Reds'
)

plt.title("SC-Params Risk Decision Matrix\n(Credit Score vs Delayed Payments)")
plt.xlabel("Credit Score")
plt.ylabel("Delayed Payment Level")
plt.show()
```

##### 1. Why did you pick the specific chart?

This **SC-Params Risk Decision Matrix** was chosen because it visually represents **decision-making logic** rather than a simple **data distribution**. Unlike standard **bar charts** or **box plots**, this matrix maps multiple parameters simultaneously—**Credit Score** and **Delayed Payment Level**—to understand how **combined conditions influence customer risk**.

This type of visualization closely mirrors how **financial institutions** design **credit policies**, **underwriting engines**, and **risk scorecards**, making it highly **business-oriented** and **easy to explain to stakeholders**. The **heatmap format** further enhances interpretability by using **color intensity** to highlight **risk concentration** across different parameter combinations, allowing high-risk customer segments to be identified quickly and effectively.

##### 2. What is/are the insight(s) found from the chart?

From the matrix, it is clearly observed that **risk concentration increases sharply with higher delayed payment levels**, especially for customers with **Standard** and **Poor credit scores**. The most critical risk zone appears in the **High Delay + Standard Credit Score** segment, with **36,941 customers**, followed by **High Delay + Poor Credit Score** with **22,886 customers**. These segments represent customers who may appear **average on paper** due to their Standard credit score but exhibit **severe repayment issues**, making them **hidden high-risk profiles**.

In contrast, customers with **Good credit scores** are more concentrated in the **Low Delay** and **Medium Delay** bands, indicating **better repayment discipline** and **greater financial stability**. This matrix clearly demonstrates that **delayed payment behavior** can be a **stronger indicator of risk** than **credit score alone** when determining overall customer risk severity.

Yes, the insights from this **SC-Params Risk Decision Matrix** have a very strong **positive business impact**.
